### 1.8 : PREPROCESSING (LABEL-ENCODING)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.dtypes
```

Converting Categorical Data To Numerical Data
Columns names are

4. ExerciseAngina

```python
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
df['ChestPainType'] = le.fit_transform(df['ChestPainType'])
```

```python
df['RestingECG'] = le.fit_transform(df['RestingECG'])
```

```python
df['ExerciseAngina'] = le.fit_transform(df['ExerciseAngina'])
```

```python
df['ST_Slope'] = le.fit_transform(df['ST_Slope'])
```

### 1.9 : SEPARATING INTO FEATURES AND TARGET COLUMNS

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

### 1.10 : SPLITING THE DATA INTO (TRAINING AND TESTING)

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42,stratify=y)
```

### 1.11 : PREDICING BY USING (LOGISTIC REGRESSION)

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
cr=classification_report(y_pred,y_test)
print(cr)

ac = accuracy_score(y_test,y_pred)
print('Accuracy Score : ',ac)
```

We have achieved an Accuracy of 86% which is Good

### 1.12 : PREDICTING BY USING DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
Decision_T = DecisionTreeClassifier()
```

```python
Decision_T.fit(X_train,y_train)
```

```python
Decision_ypred = Decision_T.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Decision_AS = accuracy_score(y_test,Decision_ypred)
print('Accuracy Score : ',Decision_AS)
```

### 1.12 : Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
nb.fit(X_train,y_train)
```

```python
nb_yperd = nb.predict(X_test)
```

```python
print(classification_report(y_test,nb_yperd))
```
